## Q21.What are the assumptions for applying the Central Limit Theorem)

ANS.The Central Limit Theorem (CLT) states that, under certain conditions, the sampling distribution of the sample mean (or sum) of a large number of independent, identically distributed random variables will be approximately normally distributed, regardless of the original distribution of the data.

Here are the key assumptions for applying the CLT:
1. Independence:

The sampled observations must be independent.

This means that the value of one observation does not influence or provide information about another.

In practice, this is often ensured by random sampling or sampling with replacement.

2. Identically Distributed (or Similar Distributions):

Ideally, the data should come from a population with the same distribution (identical distribution).

In some variants (Lindeberg or Lyapunov CLT), identically distributed is relaxed, but the variance must be finite and certain conditions must still hold.

3. Sample Size:

The sample size must be sufficiently large.

A common rule of thumb is:

 for moderately skewed distributions.

Larger samples are needed for highly skewed or non-normal populations.

For normal populations, the sample size can be small.
4. Finite Mean and Variance:

The population from which samples are drawn should have a finite mean (Œº) and finite standard deviation (œÉ).

The CLT does not hold for distributions with infinite variance (e.g., Cauchy distribution).


## Q22.What is the concept of expected value in a probability distribution?

ANS.The expected value (also called the mean or mathematical expectation) of a probability distribution is a measure of the central tendency‚Äîthe long-run average outcome of a random variable over many repetitions of an experiment.
Definition:

For a discrete random variable  with possible values  and corresponding probabilities , the expected value is calculated as:

E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)

For a continuous random variable  with probability density function (pdf) , the expected value is:

E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx

Interpretation:

It represents the average value you would expect if the random process were repeated a large number of times.

It does not guarantee that the expected value is a possible outcome‚Äîjust the theoretical average.
Example (Discrete):

Suppose you roll a fair six-sided die. The expected value  is:

E(X) = 1\cdot\frac{1}{6} + 2\cdot\frac{1}{6} + 3\cdot\frac{1}{6} + 4\cdot\frac{1}{6} + 5\cdot\frac{1}{6} + 6\cdot\frac{1}{6} = \frac{21}{6} = 3.5

Even though you can never roll a 3.5, this is the average result over many rolls.

üîπ Use Cases:

Decision making under uncertainty (e.g., expected profit/loss)

Game theory and economics

Risk analysis

Insurance and finance models.


## Q23.How does a probability distribution relate to the expected outcome of a random variable?

ANS.A probability distribution describes how the values of a random variable are distributed ‚Äî that is, how likely each possible value is to occur. The expected outcome (or expected value) of a random variable is a key summary measure that represents the average value you would expect to see if the experiment were repeated many times.
How They Relate:

1. The Expected Value Is Calculated From the Probability Distribution:

For a discrete random variable , with possible values  and corresponding probabilities , the expected value  is:
E(X) = \sum_{i=1}^{n} x_i \cdot P(X = x_i)

E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx

2. The Distribution Shapes the Expected Value:

The values that are more likely (i.e. have higher probability) contribute more to the expected value.

This makes the expected value a weighted average, where weights come from the probability distribution.
3. The Expected Value Summarizes the Distribution:

It gives a central tendency of the distribution, though it's not always a value the variable can actually take.

For example, the expected value of a die roll (1 through 6) is 3.5 ‚Äî not a possible outcome, but still a meaningful average.
Example:

Suppose a random variable  represents the winnings from a game, with this probability distribution:

Value ()	Probability

0	0.2
10	0.5
50	0.3


Then the expected value is:

E(X) = 0 \cdot 0.2 + 10 \cdot 0.5 + 50 \cdot 0.3 = 0 + 5 + 15 = 20

So, on average, you‚Äôd expect to win ‚Çπ20 per game if you played many times.

```python
# Q2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

#ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for the discrete uniform distribution
low = 1   # inclusive
high = 7  # exclusive (like a fair 6-sided die: 1 to 6)

# Define the distribution
dist = randint(low, high)

# Define the range of x-values (possible outcomes)
x = np.arange(low, high)

# Compute the PMF for each x-value
pmf_values = dist.pmf(x)

# Plotting the PMF
plt.figure(figsize=(8, 5))
plt.stem(x, pmf_values, basefmt=" ")
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

#ANS.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1).
    p (float): The probability of success (0 <= p <= 1).

    Returns:
    float: The probability P(X = x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p*x * (1 - p)*(1 - x)
```

```python
# Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution?

# ANS.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1   # minimum value
b = 6   # maximum value (like a fair die)

# Generate values
x = np.arange(a, b + 1)
# PMF for discrete uniform is 1 / (b - a + 1)
pmf = np.ones_like(x) / (b - a + 1)
# Compute CDF
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.scatter(x, cdf, color='red', zorder=5)
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q9.Write a Python function to calculate Z-scores from a dataset¬†and¬†plot¬†them?

#ANS.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot a histogram of Z-scores.

    Parameters:
        data (list or np.ndarray): A list or array of numerical values.

    Returns:
        z_scores (np.ndarray): Array of Z-scores corresponding to input data.
    """
    # Convert input to NumPy array
    data = np.array(data)

    # Compute mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Check for zero standard deviation
    if std == 0:
        raise ValueError("Standard deviation is zero; cannot compute Z-scores.")

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    sns.histplot(z_scores, bins=20, kde=True, color='skyblue', edgecolor='black')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

   ¬†return z_scores
```

```python
# Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal¬†distribution.

# ANS.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def demonstrate_clt(dist_size=10000, sample_size=30, num_samples=1000):
    # Step 1: Generate data from a non-normal distribution (Exponential)
    population = np.random.exponential(scale=2.0, size=dist_size)

    # Step 2: Draw random samples and compute sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Step 3: Plot original distribution and sampling distribution
    plt.figure(figsize=(14, 6))

    # Plot the original (non-normal) distribution
    plt.subplot(1, 2, 1)
    sns.histplot(population, kde=True, bins=50, color='skyblue')
    plt.title("Original Exponential Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Plot the sampling distribution of the mean
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, kde=True, bins=30, color='salmon')
    plt.title(f"Sampling Distribution of the Mean\n(Sample size = {sample_size}, Samples = {num_samples})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Run the demonstration
demonstrate_clt()
```

```python
# Q12.Write a Python function to calculate and plot the standard normal distribution (mean¬†=¬†0,¬†std¬†=¬†1)5?

#ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Plots the standard normal distribution (mean = 0, std = 1).
    """
    # Generate values from -4 to 4 (covers most of the standard normal curve)
    x = np.linspace(-4, 4, 1000)

    # Calculate the standard normal PDF values
    y = norm.pdf(x, loc=0, scale=1)

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.title('Standard Normal Distribution (mean = 0, std = 1)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function
plot_standard_normal_distribution()
```

```python
# Q13.Generate random variables and calculate their corresponding probabilities using the binomial distribution5?

#ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10       # Number of trials
p = 0.5      # Probability of success

# Generate random variables (number of successes)
size = 1000
random_vars = np.random.binomial(n=n, p=p, size=size)

# Count frequency of each outcome
values, counts = np.unique(random_vars, return_counts=True)

# Normalize to get empirical probabilities
empirical_probs = counts / size

# Calculate theoretical binomial probabilities
k_values = np.arange(0, n + 1)
theoretical_probs = binom.pmf(k_values, n, p)

# Print results
print("Random Variables (first 20):", random_vars[:20])
print("Empirical Probabilities:", dict(zip(values, empirical_probs)))
print("Theoretical Probabilities:", dict(zip(k_values, theoretical_probs)))

# Optional: Plot empirical vs theoretical probabilities
plt.figure(figsize=(10, 6))
plt.bar(values - 0.2, empirical_probs, width=0.4, label='Empirical', color='skyblue')
plt.bar(k_values + 0.2, theoretical_probs, width=0.4, label='Theoretical', color='lightcoral')
plt.xticks(k_values)
plt.xlabel("Number of Successes (k)")
plt.ylabel("Probability")
plt.title(f"Empirical vs. Theoretical Binomial Distribution (n={n}, p={p})")
plt.legend()
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution5?

# ANS.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_score(x, mean, std):
    """Calculate the Z-score for a data point."""
    return (x - mean) / std

def plot_standard_normal_with_z(z_score):
    """Plot the standard normal distribution and mark the Z-score."""
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.fill_between(x, 0, y, where=(x <= z_score), color='red', alpha=0.2)

    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example values
data_point = 75
population_mean = 70
population_std = 5

# Calculate Z-score
z = calculate_z_score(data_point, population_mean, population_std)
print(f"The Z-score for data point {data_point} is: {z:.2f}")

# Visualize
plot_standard_normal_with_z(z)
```

```python
# Q15.  Implement hypothesis testing using Z-statistics for a sample¬†dataset5?

# ANS.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = population_std / np.sqrt(n)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Two-tailed test: find p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision
    reject_null = p_value < alpha

    # Print results
    print("Sample Mean:", sample_mean)
    print("Z-Score:", z_score)
    print("P-Value:", p_value)
    print("Reject Null Hypothesis:" if reject_null else "Fail to Reject Null Hypothesis")

    return z_score, p_value, reject_null
    # üî¢ Example data
sample_data = [100, 102, 98, 101, 99, 97, 103, 100, 104, 98]
population_mean = 100
population_std = 2  # Known population standard deviation

# üîç Perform Z-test
z_test(sample_data, population_mean,population_std)
```

```python
# Q16.Create a confidence interval for a dataset using Python and interpret¬†the¬†result5?

# ANS.

import numpy as np
import scipy.stats as stats

# Sample data
data = [45, 48, 50, 47, 49, 52, 46, 51, 53, 48]

# Step 1: Calculate the sample statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
confidence = 0.95

# Step 2: Calculate the standard error
sem = std_dev / np.sqrt(n)

# Step 3: Get the t-critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Step 4: Compute the margin of error
margin_of_error = t_critical * sem

# Step 5: Compute the confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```

```python
# Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval¬†for¬†its¬†mean5?

# ANS.

import numpy as np
import scipy.stats as stats

# 1. Generate data from a normal distribution
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# 2. Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)
standard_error = sample_std / np.sqrt(n)

# 3. Compute 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf((1 + confidence_level) / 2)  # z-score for 95% CI ~1.96
margin_of_error = z_critical * standard_error

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# 4. Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q19.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson¬†distribution5?

# ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_val = 4  # mean (Œª) of Poisson distribution

# Values (discrete)
x = np.arange(0, 15)

# CDF Calculation
cdf_values = poisson.cdf(x, mu=lambda_val)

# Print CDF values
for xi, cdf in zip(x, cdf_values):
    print(f"P(X ‚â§ {xi}) = {cdf:.4f}")

# Plotting
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='post', label=f'Poisson CDF (Œª={lambda_val})')
plt.xlabel('x')
plt.ylabel('CDF: P(X ‚â§ x)')
plt.title('Cumulative Distribution Function - Poisson Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q23.Implement Z-score normalization on a dataset and visualize its transformation5?

# ANS.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate synthetic dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Normal distribution with mean=50, std=15

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Plot original vs normalized data
plt.figure(figsize=(12, 6))

# Original data
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True, bins=30, color='skyblue')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normalized data
plt.subplot(1, 2, 2)
sns.histplot(z_scores, kde=True, bins=30, color='lightgreen')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Q24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution?

# ANS.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(size=1000, mean=0, std_dev=1):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
        size (int): Number of samples to generate.
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.

    Returns:
        dict: A dictionary containing skewness and kurtosis.
    """
    # Generate normal distribution data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # By default, Fisher's definition (normal => 0)

    return {
        'Skewness': data_skewness,
        'Kurtosis': data_kurtosis
    }

# Example usage:
results = analyze_normal_distribution_skew_kurtosis()
print("Skewness:", results['Skewness'])
print("Kurtosis:", results['Kurtosis'])
```
